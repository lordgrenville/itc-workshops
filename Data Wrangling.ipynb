```python
import yaml
import pandas as pd
```

```python
def get_yaml(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        YAML = yaml.load(f)
    print ('target_column', YAML['target_column'])
    print ('category_column', YAML['category_column'])
    return YAML
```

```python
from glob import glob
```

```python
for filename in glob('*/*.yml'):
    print (filename)
    with open(filename, 'r', encoding='utf-8') as f:
        YAML = yaml.load(f)
        for key in YAML:
            print (key, YAML[key])
        print ('*************************')
```

## VIA

```python
df = pd.read_csv('VIA/via_morning_requests_updated_weather_shifted.csv', index_col=0); df.head()
```

```python
YAML = get_yaml('VIA/Problem definition_Scenario_1.yml')
```

```python
names = YAML['supporting_columns']
df_categorical = df.copy()
df_categorical.set_index('date', inplace=True)
df_categorical = pd.concat([pd.get_dummies(df_categorical[YAML['category_column'][0]]), df[names].set_index(df_categorical.index), 
               df_categorical[YAML['target_column'][0]]], axis=1)
df_categorical.head()
```

```python
df_categorical.to_csv('VIA/data_Scen_1.csv')
```

```python
YAML = get_yaml('VIA/Problem definition_Scenario_2.yml')

names = YAML['supporting_columns']
df_categorical = df.copy()
df_categorical.set_index('date', inplace=True)
df_categorical = pd.concat([pd.get_dummies(df_categorical[YAML['category_column'][0]]), df[names].set_index(df_categorical.index), 
               df_categorical[YAML['target_column'][0]]], axis=1)
df_categorical.head()
```

```python
df_categorical.to_csv('VIA/data_Scen_2.csv')
```
